```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sbn
```

```python
df = pd.read_csv('trees.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# Gerando o histograma com o numpy
h = np.histogram(df.iloc[:,1], bins=6) # Pega todas as linhas da coluna "Altura", com 6 eixos (bins)
h
```

```python
plt.hist(df.iloc[:, 1], bins=6)
plt.title("Árvores")
plt.ylabel("Frequência")
plt.xlabel("Altura")
```

```python
# Histograma com 6 divisões
plt.hist(df.iloc[:, 1], bins=6)
```

```python
# Histograma sem linha de distribuição de frequência com seaborn
sbn.histplot(df.iloc[:,1], kde=False, color="royalblue").set(title="Árvores") # Kde = linha de densidade
```

```python
# Método Kde plot para densidade
sbn.kdeplot(df.iloc[:, 1], color="royalblue").set(title="Árvores")
```

```python
# Densidade e histograma
sbn.histplot(df.iloc[:, 1], kde=True, bins=6, color="orangered").set(title="Árvore")
```

```python
# Gráfico de dispersão no plt
plt.scatter(x=df.Girth, y=df.Volume, color="blue", facecolors="none", marker=".")
plt.title("Árvores")
plt.ylabel("Volume")
plt.xlabel("Circunferência")

# OBS: se removermos o marker, teremos um gráfico de linhas
```

```python
# Gráfico de dispersão com seaborn
sbn.regplot(x=df.Girth, y=df.Volume, color="royalblue", data=df, x_jitter=0.3, fit_reg=False) # Fit_reg = linha de tendência
'''
O problema do gráfico de dispersão é a sobreposição dos dados.
O jitter serve para adicionarmos aleatoriedade aos dados, melhorando a visualização
'''
```

```python
# Subgráficos e divisão de telas
plt.scatter(x=df.Girth, y=df.Volume)
```

```python
plt.scatter(df.Girth, df.Height)
```

```python
plt.scatter(df.Height, df.Volume)
```

```python
plt.hist(df.Volume)
```

```python
# Small multiples
# Primeiro geramos uma figura para posicionarmos nossos elementos
plt.figure(1)
plt.subplot(2,2,1) # 2 linhas, 2 colunas, posição 1
plt.scatter(df.Girth, df.Volume)
plt.subplot(2,2,2)
plt.scatter(df.Girth, df.Height)
plt.subplot(2,2,3)
plt.scatter(df.Height, df.Volume)
plt.subplot(2,2,4)
plt.hist(df.Volume)
```

```python
# Boxplot
plt.boxplot(df.Volume, vert=False, showfliers=True, notch=False, patch_artist=True)
plt.title("Volume")
plt.xlabel("Volume")
```

```python
# Boxplot com seaborn
sbn.boxplot(data=df.Volume, orient="v").set_title("Árvores")
```

```python
# Histograma com seaborn
# 10 divisões
# Gráfico de densidade (kernel density - suavização)
# stat: count, frequency, probability, percent, density
sbn.histplot(data=df.Volume, color="indigo", kde=True, stat="density", bins=10).set_title("Árvores")
```
