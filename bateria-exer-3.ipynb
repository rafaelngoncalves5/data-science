```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sbn
```

```python
df = pd.read_csv('dados.csv', sep=';')
```

```python
df.head()
```

```python
# Renomeando VALOREMPENHO
df.rename(columns={"VALOREMPENHO": "EMPENHO"}, inplace=True)
```

```python
# Empenho
sbn.histplot(df.EMPENHO, kde=True, bins=10, color="indigo").set(title="Empenho")
```

```python
# PIB
sbn.histplot(df.PIB, kde=True, bins=10, color="teal").set(title="PIB")
```

```python
top10_pib = df.sort_values("PIB").head(10)
top10_pib
```

```python
pib_por_municipio = top10_pib.groupby(top10_pib["MUNICIPIO"])["PIB"].sum()
```

```python
pib_por_municipio.plot.bar(color="pink")
plt.title("PIB por Município")
```
