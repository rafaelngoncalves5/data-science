```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sbn
from mpl_tookits.mplot3d import axes3d
```

```python
df = pd.read_csv('co2.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
x = df.conc
y = df.uptake
```

```python
# Retorna os valores únicos do atributo "Treatment"
unicos = list(set(df.Treatment))
unicos
```

```python
for i in range(len(unicos)):
    indice = df.Treatment == unicos[i]
    plt.scatter(x[indice], y[indice], label=unicos[i])

plt.legend(loc="lower right")
```

```python
df = pd.read_csv('insect.csv')
```

```python
df.head()
```

```python
# Agrupando por spray, contando e somando os registros
by_spray = df.groupby(['spray'])['count'].sum()
by_spray
```

```python
# Gráfico de barras
by_spray.plot.bar(color="royalblue")
```

```python
by_spray.plot.bar(color=('blue', 'red', 'g', '#999999', 'pink', 'orange'))
```

```python
# Gráfico de pizza
by_spray.plot.pie()
```

```python
by_spray.plot.pie(legend=True)
```

```python
# Gráfico de dispersão com seaborn
sbn.scatterplot(x=df.conc, y=df.uptake, hue=df.Type, data=df)
```

```python
quebec = df.loc[df['Type'] == "Quebec"]
mississippi = df.loc[df['Type'] == "Mississippi"]
```

```python
plt.figure()
plt.subplot(2,1,1)
sbn.scatterplot(x=quebec.conc, y=quebec.uptake).set_title('Quebec')
plt.subplot(2,1,2)
sbn.scatterplot(x=mississippi.conc, y=mississippi.uptake).set_title('Mississippi')
plt.tight_layout()
plt.show()
```
