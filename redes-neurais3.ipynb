```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
```

```python
df = pd.read_csv('Credit2.csv', sep=';')
df
```

```python
# Separação das variáveis, ignora primeiro pois não tem valor semântico
x = df.iloc[:, 1:10].values
y = df.iloc[:,10].values
x
```

```python
# Label ecoder coluna checking_status
# Atribui valores de zero a 3
labelencoder = LabelEncoder()
x[:,0] = labelencoder.fit_transform(x[:,0])
x
```

```python
# One hot encoder coluna credit_history
# Deve adicionar 5 colunas
onehotencoder = make_column_transformer((OneHotEncoder(categories='auto'), [1]), remainder='passthrough')
x = onehotencoder.fit_transform(x)
x
```

```python
# Excluímos a variável para evitar dummy variable trap
x = x[:,1:]
x
```

```python
# Label encoder com a classe
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
y
```

```python
# Separação em treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
print(len(x_train), len(x_test), len(y_train), len(y_test))
```

```python
# Feature Scalling, Padronização z-score
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_test
```

```python
classifier = Sequential()
classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu', input_dim=12))
classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))
classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
classifier.fit(x_train, y_train, batch_size=10, epochs=100)
```

```python
y_pred = classifier.predict(x_test)
y_pred = (y_pred > 0.5)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```
