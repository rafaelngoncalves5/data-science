```python
# Verificar se os dados estão
from scipy import stats
from scipy.stats import norm, skewnorm
from matplotlib import pyplot as plt
```

```python
# Gerando nossa base normalmente distribuída
df = norm.rvs(size=1000)
df.shape
```

```python
# Método 1: Histograma
plt.hist(df, bins=20)
plt.title("Distribuição Normal")
```

```python
# Método 2: qqplot
fig, ax = plt.subplots()
stats.probplot(df, fit=True, plot=ax)
plt.title("Dist")
```

```python
# Método 3: teste de hipótese
stats.shapiro(df)
# Se o 2º valor for > 0.05, não teremos elementos para rejeitar a hipótese nula
```

```python
# Dados não normalmente distribuídos
df2 = skewnorm.rvs(4, size=1000)
```

```python
plt.hist(df2, bins=20)
plt.title("Não Normais")
plt.show()
```

```python
fig2, ax2 = plt.subplots()
stats.probplot(df2, fit=True, plot=ax2)
plt.title("nDist")
```

```python
stats.shapiro(df2)
```
