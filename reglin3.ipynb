```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from statistics import stdev
```

```python
df = pd.read_csv('slr12.csv', sep=';')
df.shape
```

```python
df.head()
```

```python
# Criando nossas variáveis dependentes e independentes
x = df.iloc[:,1].values
y = df.iloc[:,0].values
cor = np.corrcoef(x,y)
```

```python
type(x)
```

```python
# Como está em formato de ndarray, precisamos converter para matriz
x = x.reshape(-1,1)
x
```

```python
# Criando o modelo
model = LinearRegression()
model.fit(x,y)
```

```python
model.coef_
```

```python
# Plotar em gráfico
plt.scatter(x,y)
plt.plot(x, model.predict(x), color='r')
```
