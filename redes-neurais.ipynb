```python
# Redes neurais artificiais com Keras
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import numpy as np
```

```python
# Carregando a base de dados e criação dos previsores (variáveis independentes - x) e da classe (variável dpeendente - y)
df = datasets.load_iris()
classe = df.target
previsores = df.data
classe
```

```python
# Transformação da classe para formato dummy, pois temos um rede neural com três neurônios na camada de saída
classe_dummy = to_categorical(classe)
classe_dummy
```

```python
# Divisão da base de dados entre treinamento e teste (30/70)
x_treinamento,x_teste,y_treinamento,y_teste = train_test_split(previsores, classe_dummy, test_size=0.3,random_state=0)
```

```python
# Criação da estrutura de rede neural com a classe Sequential (sequência de camadas)
modelo = Sequential()
# Primeira camada oculta, 5 neurônios, 4 neurônios de entrada
modelo.add(Dense(units=5,input_dim=4))
# Segunda camada oculta
modelo.add(Dense(units=4))
# Terceira camada oculta
# Função softmax porque temos um problema de classificação com mais de duas classes
# é gerada uma probabilidade em cada neurônio
modelo.add(Dense(units=3,activation='softmax'))
```

```python
# Visualização da estrutura de rede neural
modelo.summary()
```

```python
# Configuração dos parâmetros da rede neural (adam = algoritmo para atualizar os pesos e loss = cálculo de erro)
modelo.compile(optimizer='adam', loss='categorical_crossentropy',
               metrics=['accuracy'])
# Treinamento, dividindo a base de treinamento em uma porção para validação (validation_data)
modelo.fit(x_treinamento, y_treinamento, epochs=1000, validation_data = (x_teste,y_teste))
```

```python
# Previsões e mudar variável para True ou False de acordo com o treshold 0.5
previsoes = modelo.predict(x_teste)
previsoes = (previsoes > 0.5)
previsoes
```

```python
# Como é um problema de três saídas, precisamos buscar a posição que possui o maior valor (são retornados 3 valores)
y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]
```

```python
# Geração da matriz de confusão
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
confusao
```
