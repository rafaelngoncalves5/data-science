```python
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('soybean.csv')
df
```

```python
previsores = df.iloc[:,0:-1].values
classe = df.iloc[:,-1].values
```

```python
# Converte a classe para números
le = LabelEncoder()

classe[:] = le.fit_transform(classe[:])
classe
```

```python
# Converte os previsores para números
le2 = LabelEncoder()
for i in range(0,35):
    previsores[:,i] = le2.fit_transform(previsores[:,i])

# Convertendo para float32
previsores = previsores.astype('float32')
```

```python
# Converte a classe numérica para dummy
dummy_class = to_categorical(classe)
dummy_class
```

```python
# Divisão da base de teste e treino
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, dummy_class, test_size=0.3, random_state=0)
x_teste
```

```python
# Criando a rede neural
modelo = Sequential()

# Camada de entrada oculta com a quantidade de previsores (input_dim)
modelo.add(Dense(units=64, input_dim=35, activation='relu'))
# Segunda camada
modelo.add(Dense(units=32))
# Camada de saída com a quantidade de classificadores
modelo.add(Dense(19, activation='softmax'))
```

```python
modelo.summary()
```

```python
# Compilação do modelo e configuração dos híper-parâmetros
modelo.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Treinamento e teste do modelo
modelo.fit(x_treino, y_treino, epochs=1000, validation_data=(x_teste,y_teste))
```

```python
previsoes = modelo.predict(x_teste)
previsoes
```

```python
y_teste_matriz = [np.argmax(t) for t in y_teste]
y_previsao_matriz = [np.argmax(t) for t in previsoes]
```

```python
mc = confusion_matrix(y_teste_matriz, y_previsao_matriz)
mc
```
