```python
# c-means
# Não adiciona instâncias a grupos de forma absoluta
# Atribui aos cluster percentualmente cada instância
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import skfuzzy
```

```python
iris = datasets.load_iris()
```

```python
# C -> clusters definidos, iris.data.T -> matriz transposta. O resto é default
r = skfuzzy.cmeans(data = iris.data.T,c=3,m=2,error=0.005,maxiter=1000,init=None)
```

```python
# Obtendo as porcentagens de um registro pertencer a um cluster, que está na posição 1 da matriz
previsoes_p = r[1]
```

```python
for x in range(150):
    print(previsoes_p[0][x], previsoes_p[1][x], previsoes_p[2][x])
```

```python
previsoes = previsoes_p.argmax(axis=0)
resultados = confusion_matrix(iris.target, previsoes)
resultados
```
