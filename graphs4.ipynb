```python
from igraph import Graph, plot
```

```python
grafo = Graph(edges=[(0,2),(0,1),(1,4),(1,5),(2,3),(6,7),(3,7),(4,7),(5,6)],
              directed=True)
grafo.vs['label'] = ['a','b','c','d','e','f','g','h']
grafo.es['weight'] = [2,1,2,1,2,1,3,1]
```

```python
plot(grafo,bbox=(0,0,300,300),edge_label=grafo.es['weight'])
```

```python
cvert = grafo.get_shortest_paths(0,7,output='vpath')
cvert
```

```python
care = grafo.get_shortest_paths(0,7,output='epath')
care
```

```python
# Mostra o id dos vértices que fazem parte do caminho
careid = []
for n in care[0]:
    careid.append(n)
careid
```

```python
# Mostra o nome dos vértices que fazem parte do caminho
cnv = []
for n in cvert[0]:
    print(grafo.vs[n]['label'])
    cnv.append(grafo.vs[n]['label'])
cnv
```

```python
# Colorir vértices do caminho
for v in grafo.vs:
    if v['label'] in cnv:
        v['color'] = 'green'
    else:
        v['color'] = 'gray'
```

```python
# Colorir arestas do caminho
for e in grafo.es:
    if e.index in careid:
        e['color'] = 'green'
    else:
        e['color'] = 'gray'
```

```python
plot(grafo,bbox=(0,0,300,300))
```
