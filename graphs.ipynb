```python
# Grafos
from igraph import Graph
from igraph import plot
```

```python
# Definição do grafo com arestas
grafo1 = Graph(edges = [(0,1), (1,2), (2,3), (3,0)], directed=True)
# Definição do rótulo de cada vértice
grafo1.vs['label'] = range(grafo1.vcount())
print(grafo1)
```

```python
# Gráfico
plot(grafo1, bbox=(0,0,300,300))
```

```python
# Criação do segundo grafo
grafo2 = Graph(edges=[(0,1),(1,2),(2,3),(3,0),(3,2),(2,1),(1,0)], directed=True)
grafo2.vs['label'] = range(grafo2.vcount())
plot(grafo2, bbox=(0,0,400,400))
```

```python
# Grafo com laço
grafo3 = Graph(edges=[(0,1),(1,2),(2,3),(3,0),(1,1)],directed=True)
grafo3.vs['label'] = range(grafo3.vcount())
plot(grafo3, bbox=(0,0,300,300))
```

```python
# Quarto grafo
grafo4 = Graph(edges=[(0,1),(1,2),(2,3),(3,0),(1,1)],directed=True)
# Adicionando vértice isolado
grafo4.add_vertex(5)
grafo4.vs['label'] = range(grafo4.vcount())
plot(grafo4,bbox=(0,0,300,300))
```

```python
grafo5 = Graph(directed=False)
grafo5.add_vertices(10)
grafo5.add_edges([(0,1), (2,2),(2,3),(3,0)])
grafo5.vs['label'] = range(grafo5.vcount())
print(grafo5)
plot(grafo5, bbox=(0,0,300,300))
```

```python
grafo6 = Graph(directed=False)
grafo6.add_vertices(5)
grafo6.add_edges([(0,1),(1,2),(2,3),(3,4),(4,0),(0,2),(2,1)])
grafo6.add_vertex(5)
grafo6.add_vertex(6)
grafo6.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
plot(grafo6, bbox=(0,0,300,300))
```
