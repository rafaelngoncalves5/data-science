```python
from igraph import Graph
from igraph import plot
```

```python
grafo = Graph(edges=[(0,2),(0,1),(1,4),(2,3),(6,7),(3,7),(4,2),(5,6)],
              directed=True)
grafo.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
grafo.es['weight'] = [2,1,2,1,2,1,3,1]
```

```python
plot(grafo, bbox=(0,0,300,300),edge_label=grafo.es['weight'])
```

```python
# Caminho entre os vértices A - H
c_vertice = grafo.get_shortest_paths(0,7,output='vpath')
for n in c_vertice[0]:
    print(grafo.vs[n]['label'])
```

```python
c_are = grafo.get_shortest_paths(0,7,output='epath')
c_are
```

```python
# Mostrar o caminho
c_are_id = []
for n in c_are[0]:
    c_are_id.append(n)
c_are_id
```

```python
# Somatória dos pesos
dist = 0
for e in grafo.es:
    if e.index in c_are_id:
        dist += grafo.es[e.index]['weight']
dist
```
